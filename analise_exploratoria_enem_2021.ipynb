# Exploração

```python
#Total de inscritos
df['NU_INSCRICAO'].count()

#Porcentagem de inscritos por estado
print("Porcentagem de inscritos por estado:")
result = df['SG_UF_PROVA'].value_counts()/df.shape[0]*100
result.round(2)
```

```python
#quantidade de participantes por estado
fig, ax = plt.subplots(figsize=(10,5))
sns.countplot(x='SG_UF_PROVA',data=df,palette='flare',order = df['SG_UF_PROVA'].value_counts().index)
plt.xlabel('Estado')
plt.ylabel("Quantidade de inscritos")
plt.title('Gráfico: Quantidade de inscritos por Estado')
plt.ticklabel_format(style='plain', axis='y')
plt.show()
```

```python

print("Quantidade total de inscritos: ",df['NU_INSCRICAO'].count())
inscritos_tot = df['NU_INSCRICAO'].count()
```

```python
#Remove inscritos que não participaram dos dois dias de prova.
df = df.loc[((df.TP_PRESENCA_CN == 1) & (df.TP_PRESENCA_CH == 1))]
#Quantidade de inscricoes que realizaram as duas provas

inscritos_tot_final = df['NU_INSCRICAO'].count()
print('Total de inscritos que realizaram as duas provas:',df['NU_INSCRICAO'].count())
```

```python
#Abstenção
abstencao = inscritos_tot - inscritos_tot_final
print("Total de inscritos que não realizaram as duas provas:",abstencao)

#Abstenção em porcentagem
abstencao_porc = inscritos_tot_final/inscritos_tot*100
tot_porc = 100 - abstencao_porc
print("O total de abstenção foi de de ",tot_porc.round(2),"%")
```

```python
#Media da nota
#cria coluna nota média baseada na soma das 5 provas
df['NOTA_MEDIA'] = (df['NU_NOTA_CN'] + df['NU_NOTA_CH'] + df['NU_NOTA_LC'] + df['NU_NOTA_MT'] + df['NU_NOTA_REDACAO']) / 5
```

```python
#Quantidade de Parcitipantes por sexo
#Gráfico
fig, ax = plt.subplots(figsize=(10,5))
sns.countplot(x='TP_SEXO',data=df,palette='rocket_r',order = df['TP_SEXO'].value_counts().index)
plt.xlabel("Sexo")
plt.ylabel("Quantidade de participantes")
plt.title('Gráfico: Participantes por sexo')
plt.ticklabel_format(style='plain', axis='y')
plt.show()
print("Quantidade de Participantes por Sexo:")
print(df['TP_SEXO'].value_counts())
```

```python
#Gráfico - Nota Média por Sexo
sns.boxplot(x=df['TP_SEXO'], y=df['NOTA_MEDIA'],palette='rocket_r')
plt.title('Gráfico: Nota média dos participantes por Sexo')
plt.xlabel("Sexo")
plt.ylabel("Nota Média")
plt.show()
```

# Conclusão

1 – A quantidade de total de inscritos no <b>ENEM 2021 foi de 3.389.832</b>. <br><u>Já a quantidade final de
participantes que realizaram as duas provas</u>, nos dois dias de aplicação foi de <b>2.238.107</b>, sendo
assim, o ENEM 2021 teve uma abstenção de <b>33,98%</b>, totalizando <b>1.151.725</b> inscritos, que por
algum motivo não realizaram as duas provas.<br><br>
2- Os três estados com maior número de inscritos foram <b>São Paulo, Minas Gerais e Bahia</b>.<br><br>
3 – O sexo Feminino foi predominante na quantidade de participantes, com <b>1.381.949</b>, contra
856.158 participantes do sexo masculino. Já a média de notas dos participantes do sexo
masculino foi maior que a média de nota dos participantes do sexo feminino.<br><br>
4 – A cor/raça predominante entre os participantes foi Branca e Parda. As três maiores médias
de notas foram predominantes nas cores/raças <b>Branca, Amarela e Não declarada</b>.<br><br>
5 – Os números de participantes por escola são: Não respondeu com <b>1.390.710</b>, Escola Publica
com <b>668.036</b> e Escola Particular com <b>179.361</b>. Com base nos dados apresentados, entende-se
que a grande maioria dos participantes preferem não declarar em que tipo de escola estudou.<br><br>
6 – O número de participantes que possuem internet em casa é muito superior ao número que
participantes que não possuem internet em casa.<br><br>
7 – O número de participantes apresenta um grande fator decrescente conforme a idade
aumenta. Sendo o ápice das participações entre os inscritos de <b>17 a 20 anos</b>.<br>
